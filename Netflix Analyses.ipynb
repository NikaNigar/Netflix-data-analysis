```python
df['director']=df['director'].fillna('unknown')
df['director'].isnull().sum()
```

```python
df.isnull().sum()
```

```python
df['cast']=df['cast'].fillna('unknown')
df.isnull().sum()
```

```python
df['country']=df['country'].fillna(df['country'].mode()[0])
df.isnull().sum()
```

```python
df=df.dropna(subset=['date_added'])
df.isnull().sum()
```

```python
df=df.dropna(subset=['rating'])
df.isnull().sum()
```

```python
df=df.dropna(subset=['duration'])
df.isnull().sum()
```

```python
from matplotlib import pyplot as plt
```

```python
df.head()
```

```python
counts=df['type'].value_counts()
counts.to_frame()
dilimler=[0.1,0.0]
```

```python
plt.pie(counts,
        labels=counts.index,
        explode=dilimler,
        startangle=90,
        autopct='%1.0f%%',
        colors=['r','g'])

plt.title('Movies')
plt.show()
```

```python
country_counts=df['country'].value_counts()
country_counts=country_counts.to_frame().sort_values(by='count', ascending=False).head(10)
country_counts
```

```python
country_counts = df['country'].value_counts().to_frame(name='count')
country_counts = country_counts.sort_values(by='count', ascending=False).head(10)

plt.barh(country_counts.index[::-1],
         country_counts['count'][::-1])

plt.xlabel('Counts')
plt.ylabel('Countries')
plt.title('Netflix Movie Ranks')
plt.show()
```
